# Matplotlib


## What is Matplotlib?
Matplotlib is the standard visualization library for Data Science and Machine Learning.


To check more please visit: [Matplotlib](www.matplotlib.org)

## Import Matplotlib
To import matplotlib do the following:

```
import matplotlib.pyplot as plt
```

## plt.plot(x, y, color='color', linestyle='any linestyle')
This is the most basic plot command:

Check the docs: [Matplotlib Plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

```
x = [0, 1, 2]
y = [100, 200, 300]
```

```
plt.plot(x, y, color='green', linestyle='--')
```

## plt.show()
In return we get the image of our plot:

Check the docs: [Matplotlib Show](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.show.html?highlight=show#matplotlib.pyplot.show)

```
plt.plot(x, y)
plt.show()
```

## plt.scatter(x, y)
We can make a scatter plot to show the data:

Check the docs: [Matplotlib Scatter](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html?highlight=scatter#matplotlib.pyplot.scatter)

```
import pandas as pd
```

```
housing = pd.DataFrame({'rooms': [1, 1, 2, 2, 2, 3, 3, 3],
                        'price': [100, 120, 190, 200, 230, 310, 330, 305]})
housing
```

```
plt.scatter(housing['rooms'], housing['price'])
plt.show()
```

## plt.title('String'), plt.xlabel('String') and plt.ylabel('String')

*  We can add a title in our plot:  **.title()**  
*  We can add a label on the x axis: **.xlabel()**
*  We can add a label on the y axis: **.ylabel()**
We can modify the text size by saying: , size=#

Check the docs: [Matplotlib Title](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.title.html?highlight=title#matplotlib.pyplot.title)

```
plt.plot(x, y)
plt.title('Title', size=20)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.show()
```

## plt.xlim(range)
We can set the limit for the x axis values, this works with ylim():

Check the docs: [Matplotlib xlim](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlim.html?highlight=xlim)

```
plt.plot(x, y)
plt.xlim(0, 2)
plt.ylim(100, 300)
plt.show()
```

## plt.figure(figsize=(x, y))
We can adjust the graph size by providing both x and y values:
*  This command must be typed before typing .plot()

Check the docs: [Matplotlib Figure](https://matplotlib.org/stable/api/_as_gen/matplotlib.figure.Figure.html#matplotlib.figure.Figure)

```
plt.figure(figsize=(12, 8))
plt.plot(x,y)
plt.show()
```

# Seaborn

## What is Seaborn?
Seaborn is a visualization library that focuses on creating standard statistical plots with a simple one line function call.


To learn more visit: [Seaborn](https://seaborn.pydata.org/)

## Import Seaborn
To import Seaborn do the following:

```
import seaborn as sns
```

## sns.displot(df['column name'], kde=Bool, bins=#, color='color')
This will generate a histogram:
*  kde will show or hide a line in the histogram
*  bins will show a fixed amount of columns on the graph

Check the docs at: [Seaborn Displot](https://seaborn.pydata.org/generated/seaborn.displot.html?highlight=displot#seaborn.displot)

```
import pandas as pd
```

```
df = pd.read_csv('heart.csv')
df.head()
```

```
sns.displot(df['age'], kde=False, bins=18)
```

## sns.countplot(x='column name', data=df, hue='another column', palette='color style')
This will generate a histogram across a categorical variable:

*  For the palette please check all available palettes in: [Matplotlib Palettes](https://matplotlib.org/stable/tutorials/colors/colormaps.html)

Check the docs at: [Seaborn Countplot](https://seaborn.pydata.org/generated/seaborn.countplot.html?highlight=countplot#seaborn.countplot)

```
df.head()
```

```
sns.countplot(x='cp', data=df, hue='sex', palette='bone')
```

## sns.boxplot(x='categorical column', y='continuous column', data=df)
This will generate the distribution across different categories:

Check the docs: [Seaborn Boxplot](https://seaborn.pydata.org/generated/seaborn.boxplot.html?highlight=boxplot#seaborn.boxplot)

```
df.head()
```

```
sns.boxplot(x='sex', y='age', data=df)
```

## sns.scatterplot(x='column', y='column', data=df)
This will return a scatter plot with Seaborn:

Check the docs: [Seaborn Scatterplot](https://seaborn.pydata.org/generated/seaborn.scatterplot.html?highlight=scatterplot#seaborn.scatterplot)

```
df.head()
```

```
sns.scatterplot(x='chol', y='trestbps', data=df)
```

## sns.pairplot(df, hue='column variable')
This will generate multiple graphs:

Check the docs in: [Seaborn Pairplot](https://seaborn.pydata.org/generated/seaborn.pairplot.html)

```
iris = pd.read_csv('iris.csv')
iris.head()
```

```
sns.pairplot(iris, hue='species')
```
